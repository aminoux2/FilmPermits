```python
#change to date time datatype 
film['StartDateTime'] = pd.to_datetime(film['StartDateTime'])
film['EndDateTime'] = pd.to_datetime(film['EndDateTime'])
```

```python
#makes a new column for day of week and year
film['ShootDayOfWeek']= film['StartDateTime'].dt.weekday
film['ShootMonth']= film['StartDateTime'].dt.month
film['ShootYear']= film['StartDateTime'].dt.year

```

```python
film.info()
```

```python
#changing data type to categorical
film= film.astype({'ShootYear':str})
```

```python
#dropping columns
film.drop(['EventAgency'],axis=1, inplace=True)
film.drop(['Country'],axis=1, inplace=True)
```

```python
#changed to month name
film['ShootMonth'] = film['ShootMonth'].apply(lambda x: calendar.month_abbr[x])
```

```python
#changed to day of week name
film['ShootDayOfWeek'] = film['ShootDayOfWeek'].apply(lambda x: calendar.day_abbr[x])
```
